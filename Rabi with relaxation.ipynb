```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range, relaxation rate, and Rabi frequency
t = np.linspace(0, 50, 1000)
gamma = 0.05  # Relaxation rate for noticeable relaxation effects
Omega = 5.0  # Rabi frequency

# Corrected expressions for the populations with relaxation ensuring the sum is always 1
n_plus = 0.5 * (1 + np.exp(-gamma * t) * np.cos(Omega * t))
n_minus = 1 - n_plus  # Ensures total probability is always 1

# Plotting the populations over the extended time to show relaxation to equilibrium
plt.figure(figsize=(10, 6))
plt.plot(t, n_plus, label=r'$n_+ \ (\left| e \right\rangle)$', linestyle='--', color='orange')
plt.plot(t, n_minus, label=r'$n_- \ (\left| g \right\rangle)$', linestyle='-', color='blue')
plt.title('Rabi Oscillations with Relaxation')
plt.xlabel('Time (1/)')
plt.ylabel('Population')
plt.axhline(0.5, color='green', linestyle='--', label='Equilibrium at 0.5')
plt.legend()
plt.grid(True)
plt.show()
```
