```python
import numpy as np
import matplotlib.pyplot as plt

def rabi_oscillations(omega, delta, T, dt):
    """
    Simulate Rabi oscillations for a two-level system.
    
    Parameters:
    - omega: Rabi frequency (angular frequency of the driving field)
    - delta: Detuning (difference in energy between the driving field and the energy difference of the levels)
    - T: Total time for the simulation
    - dt: Time step for the simulation
    
    Returns:
    - times: Array of time points
    - probabilities: Array of probabilities for being in the excited state
    """
    times = np.arange(0, T, dt)
    probabilities = np.square(np.sin(0.5 * np.sqrt(omega**2 + delta**2) * times)) * (omega**2 / (omega**2 + delta**2))
    
    return times, probabilities

def plot_rabi_oscillations_pi_notation(omega, deltas, T, dt):
    times = np.arange(0, T, dt)
    plt.figure(figsize=(12, 6))

    # Define labels using pi notation
    labels = [f"Delta = {delta/np.pi:.2f} rad/s" if delta != 0 else "Delta = 0 rad/s" for delta in deltas]

    for delta, label in zip(deltas, labels):
        probabilities = rabi_oscillations(omega, delta, T, dt)[1]
        plt.plot(times, probabilities, label=label)

    plt.xlabel('Time (s)')
    plt.ylabel('Probability')
    plt.title('Rabi Oscillations for Detuning')
    plt.legend()
    plt.grid(True)
    plt.show()

# Parameters
omega = 2 * np.pi * 5  # Rabi frequency, 5 Hz
deltas = [0, 2 * np.pi * 2, 2 * np.pi * 5]  # Different detuning values in rad/s: 0 rad/s, 2 rad/s, and 5 rad/s
T = 2  # Total time, 2 seconds
dt = 0.01  # Time step

# Plot with updated title
plot_rabi_oscillations_pi_notation(omega, deltas, T, dt)
```
